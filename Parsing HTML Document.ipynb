```python
# BeautifulSoup (content, HTML.Parser)
# Different parsers are (html5lib, html.parser, lxml, xml)
```

```python
from bs4 import BeautifulSoup

# Reading HTML file
fh=open('Web.html','r')

# Reading content of HTml file
data=fh.read()

# Creating Soup object
soup=BeautifulSoup(data,'html.parser')
```

```python
# Getting all text

soup.get_text()
```

```python
# Getting all the contents with tags

soup.prettify()
```

```python
# Printing the tags
# soupobjects.Outertag.Innertag

print(soup.head)
print(soup.head.title)
print(soup.h1)
print(soup.h2)                 # only the first occurence of h2 tag will be returned.
print(soup.li)
print(soup.li)
print(soup.li)
```

# Finding all tags

## By find_all () method, we can find all the elements which meet some criteria. Ex: finding all li tags.

```python
print(soup.final_all("li"))
for tag in soup.final_all("li"):
    print("{0}: {1}".format(tag.name, tag.text))
```

```python
# The name attribute of a tag gives its name and the text attribute gives the text content.  

print("HTML tag: {0}, name: {1}, text: {2}".format(soup.h2,soup.h2.name,soup.h2.text))
```

```python
soup.recursiveChildGenerator()
```

```python
# With recursiveChildGenerator() method, we traverse the html document & print all the HTML tags.

for child in soup.recursiveChildGenerator():
    if child.name:
        print(child.name)
```

# Finding element by ID

## By find() method, we can find elements by various means including element id.

```python
print(soup.find("h1"))
print(soup.find("h1",id="mylist"))

print("\n")

print(soup.find("h2"))
print(soup.find("h2",id="mylist"))
```

# Appending new element

## The append() method appends a new tag to the HTML document.

```python
newtag=soup.new_tag('li')
newtag.string="Cherry"
ultag=soup.ul
ultag.append(newtag)
print(ultag.prettify())
```

# Inserting new element

## The insert() method appends a new tag to the HTML document.



```python
newtag=soup.new_tag('li')
newtag.string="Watermelon"
ultag=soup.ul
ultag.append(2,newtag)
print(ultag.prettify())
```

# Deleting the element

## The clear() method removes the content of the tag.

```python
foundtag.soup.title
foundtag.clear()
print(soup.prettify())
```

```python
foundtag.soup.title
foundtag.extract()
print(soup.prettify())
```
